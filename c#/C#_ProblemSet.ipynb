#### Implement function triangleArea(a,b,c) that takes as input the lengths of the 3 sides of a triangle and returns the area of the triangle. By Heron's formula, the area of a triangle with side lengths a, b, and c is $\sqrt{s(s-a)(s-b)(s-c)}$, where s = (a+b+c)/2. 

```
var a = 2;
var b = 2;
var c = 2;
var s = (a+b+c)/2;
double triangleArea = Math.Sqrt(s*(s-a)*(s-b)*(s-c));
Console.WriteLine($"Area of the triangle with sides '{a}','{b}' and '{c}' is {triangleArea} units");
```

### Question 4
#### Write a program in C# Sharp to separate odd and even integers in separate arrays

```
int[] array = {25,47,42,56,32}; // Enter the numbers to be seperated as odd and even in this array.
int[] even = new int[10]; // Array size need to be defined. Here, I have taken it to be greater than the number of total numbers inputted.
int[] odd = new int[10];
int i=0;
int j=0;
int k=0;
int n;

for(i=0; i<array.Length; i++)
{
    if (array[i]%2 == 0) // Each number in the array is divided by 2 and the division remainder checked whether equal to 0.
    {
        even[j] = array[i]; // When the remainder is 0, that number is stored in to a new array - 'even'.
        j++; // Index of the array is incremented by 1 to store the next value.
    }
    else
    {
        odd[k] = array[i]; // When the remainder is not equal to 0, that number is stored in to a new array - 'odd'.
        k++;
        }
}

Console.WriteLine("Even numbers: ");
for(i=0; i<j; i++)
{
    Console.WriteLine($"{even[i]} ");
}

Console.WriteLine(" \n ");

Console.WriteLine("Odd numbers: ");
for(i=0; i<k; i++)
{
    Console.WriteLine($"{odd[i]} ");
}
```
